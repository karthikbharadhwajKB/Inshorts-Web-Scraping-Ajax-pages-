## Method for Extracting News Headlines 

```python
Headlines = [] 
def get_headlines(req):
    soup = bs4.BeautifulSoup(req,'lxml')
    headlines = soup.find_all('span',attrs={"itemprop": "headline"})
    for head in headlines:
        Headlines.append(head.text)
        print(head.text,"\n")
```

## Method for Extracting News Body

```python
article_body = []
def get_articles_body(req):
    soup = bs4.BeautifulSoup(req,'lxml')
    bodies = soup.find_all('div',attrs={"itemprop": "articleBody"})
    for body in bodies:
        article_body.append(body.text)
        print(body.text,"\n")
```

## Load More

```python
url = 'https://inshorts.com/en/ajax/more_news'
```

```python
news_offset = "nktifaez-1"
```

```python
while True:
    response = requests.post(url,data={"category":"","news_offset":news_offset})
    if response.status_code != 200:
        print(response.status_code)
        break
    response_json = json.loads(response.text)
    get_headlines(response_json['html'])
    get_articles_body(response_json['html'])
    news_offset = response_json["min_news_id"]
```
